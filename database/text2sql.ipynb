# Text2SQL

````python
from langchain_groq import ChatGroq
from dotenv import load_dotenv
from os import environ
````

````python
load_dotenv()
api_key_groq = environ['API_GROG']
````

````python
llm = ChatGroq(
    model="openai/gpt-oss-120b",
    temperature=0,
    max_tokens=500,
    api_key=api_key_groq
)
````

````python
def generate_sql_query(query, schema):
    prompt = f"""
    Given the following database schema:
    {schema}

    Translate the following query into SQL:
    "{query}"

    Return only the SQL ready to execute. Do not add anything not part of the SQL.
    """
    
    response = llm.invoke(prompt)
    response = response.model_dump()
    
    generated_sql = response["content"].strip().replace("```sql", "").replace("```", "")
    reasoning_sql = response.get("additional_kwargs", {}).get("reasoning_content", "")
    return generated_sql, reasoning_sql, response


schema = """
Table: hotels
Columns:
    id INTEGER PRIMARY KEY
    name TEXT
    address TEXT
    price INTEGER
    free_cancellation BOOL

Table: flight_schedules
Columns:
    id INTEGER PRIMARY KEY
    airplane_id INTEGER
    departure_time TIME
    arrival_time TIME
    price INTEGER
)

Table: flights
Columns:
    airplane_id INTEGER PRIMARY KEY
    airplane_name TEXT
    cabin_kg INTEGER
    baggage_kg INTEGER
    beverages TEXT (Free, Paid, Not Available)
    entertainment TEXT (Free, Paid, Not Available)
"""

query = "List down up to 5 flights departing before 17.00 with free beverages."

generated_sql, reasoning_sql, response = generate_sql_query(query, schema)
print(generated_sql)
````
