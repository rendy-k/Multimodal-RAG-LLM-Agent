# Few Shot Prompting

```python
from langchain_core.prompts import FewShotChatMessagePromptTemplate
```

```python
# Set examples
examples = [
    {"input": "The sky is dark", "output": "Langitnya gelap"},
    {"input": "The water is boiling", "output": "Airnya sedang mendidih"},
]

example_prompt = ChatPromptTemplate.from_messages(
    [
        ("human", "{input}"),
        ("ai", "{output}"),
    ]
)

# Create the few shot prompt
few_shot_prompt = FewShotChatMessagePromptTemplate(
    examples=examples,
    example_prompt=example_prompt,
    input_variables=["input"],
)

translator = ChatPromptTemplate.from_messages(
    [
        ("system", "You are an English-Indonesian translator."),
        few_shot_prompt,
        ("human", "{user_input}"),
    ]
)

chat_messages = translator.format_messages(
    user_input="Fried noodle is people's favorite food."
)

response = llm_model.invoke(chat_messages)
print(response.content)
```

# Chain

```python
from langchain_core.output_parsers import StrOutputParser
from langchain_core.runnables import RunnablePassthrough, RunnableLambda, RunnableParallel, RunnableBranch
from operator import itemgetter
```

```python
# Simple chain
prompt = ChatPromptTemplate.from_messages(
    [
        ("system", "You are an environmental scientist. Answer less than 30 words. Casual tone."),
        ("human", "{query}"),
    ]
)

str_output_parser = StrOutputParser()

simple_chain = prompt | llm_model | str_output_parser

response = simple_chain.invoke({"query": "What is air pollution?"})
print(response)
```

1. query: What is air pollution?
2. prompt: ("human", "What is air pollution?)
3. llm_model: AIMessage(content=Air pollution is basically ...
4. str_output_parser: Air pollution is basically ...

```python
# RunnablePassthrough
pass_through = RunnablePassthrough()
response = pass_through.invoke("What is water pollution?")
response
# The output is the same as the input
```

```python
def state_interest_rate(rate: float) -> float:
    return f"The interest rate is {rate}."

lambda_function = RunnablePassthrough() | RunnableLambda(state_interest_rate)
response = lambda_function.invoke("12%")
response
```

```python
# RunnableParallel
parralel = RunnableParallel(
    {
        "pass_through": RunnablePassthrough(),
        "sentence": RunnableLambda(state_interest_rate)
    }
)
response = parralel.invoke("12%")
response
```

```python
template = """Answer the question based only on the following context:
{context}

Question: {question}

Answer in {language} language in maximum 30 words.
"""

prompt_template = ChatPromptTemplate.from_template(template)

chain = (
    {
        "context": itemgetter("interest_rate") | RunnableLambda(state_interest_rate),
        "question": itemgetter("question"),
        "language": itemgetter("language"),
    } # This is a Runnable Parallel
    | prompt_template
    | llm_model
    | StrOutputParser()
)

response = chain.invoke(
    {
        "interest_rate": "12",
        "question": "If the capital is $100, how much is the return?",
        "language": "Indonesian"
    }
)
response
```
